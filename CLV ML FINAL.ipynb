3. **Working Mechanism of Evaluation Metrics**:
   - **RMSE**: It calculates the square root of the average squared differences between the predicted and actual values. It's sensitive to large errors, making it relevant for scenarios where large prediction errors are more detrimental.
   - **MAE**: This metric averages the absolute differences between predicted and actual values. It's less sensitive to outliers compared to RMSE, providing a more robust error measure in the presence of anomalies.
   - **R-squared**: This measures the proportion of variance in the dependent variable that is predictable from the independent variables. A higher R-squared indicates a better fit of the model to the data.
   - **MAPE (Mean Absolute Percentage Error)**: MAPE calculates the average of the absolute percentage errors between the predicted and actual values. It expresses accuracy as a percentage, which makes it very intuitive and easy to interpret.

In order to handle the outliers there will be 2 steps which will be performed on the dataset, they are:
* **Log Transformation**
  * For Monthly Premium Auto & Total Claim Amount
* **Winsorization**
  * For Monthly Premium Auto & Total Claim Amount (after Winsorization) & Customer Lifetime Value

Below are the benefits: 

1. **Benefits of Log Transformation**:
   - **Reduces Skewness**: It's particularly effective for right-skewed (or positively skewed) distributions. By applying a log transformation, the data becomes more normally distributed, which is a common assumption in many statistical models.
   - **Diminishes the Impact of Outliers**: Outliers have less influence post-log transformation, as the transformation compresses large values more than smaller values.
   - **Improves Model Performance**: Many models perform better when the features are normally distributed.
   - **Linear Relationships**: Log transformation can help in linearizing relationships between variables, which is beneficial for models assuming linearity.

2. **Benefits of Winsorization**:
   - **Limits Extreme Values**: By capping the data at specific percentile thresholds, it effectively reduces the impact of extreme outliers.
   - **Preserves Data Integrity**: Unlike removing outliers, winsorization modifies them, thus maintaining the size of the dataset.
   - **Enhanced Model Stability**: Models become less sensitive to the extremes of the data, leading to potentially more stable predictions.
   - **Balanced Distribution**: It can help in achieving a more symmetric distribution of the data.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to apply log transformation and handle zeros/negatives
def log_transform(column):
    # Replace zero or negative values with a small positive value (e.g., the smallest non-zero value in the column)
    min_positive = column[column > 0].min()
    column_transformed = column.replace(0, min_positive).replace(np.nan, min_positive)
    column_transformed = column_transformed[column_transformed > 0]  # Ensure no negative values
    return np.log(column_transformed)

# Applying log transformation
df['Monthly Premium Auto Log'] = log_transform(df['Monthly Premium Auto'])
df['Total Claim Amount Log'] = log_transform(df['Total Claim Amount'])

# Visualizing the distribution post-transformation
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df['Monthly Premium Auto Log'], bins=30, ax=ax[0], kde=True)
ax[0].set_title('Log Transformed Monthly Premium Auto')

sns.histplot(df['Total Claim Amount Log'], bins=30, ax=ax[1], kde=True)
ax[1].set_title('Log Transformed Total Claim Amount')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt
import seaborn as sns

# Function to apply winsorization
def apply_winsorization(column):
    return winsorize(column, limits=[0.05, 0.05])

# Applying winsorization to the log-transformed features
df['Monthly Premium Auto Log Winsorized'] = apply_winsorization(df['Monthly Premium Auto Log'])
df['Total Claim Amount Log Winsorized'] = apply_winsorization(df['Total Claim Amount Log'])

# Applying winsorization to the 'Customer Lifetime Value' column
df['Customer Lifetime Value Winsorized'] = apply_winsorization(df['Customer Lifetime Value'])

# Visualizing the distribution post-winsorization
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(df['Monthly Premium Auto Log Winsorized'], bins=30, ax=ax[0], kde=True)
ax[0].set_title('Winsorized Monthly Premium Auto Log')

sns.histplot(df['Total Claim Amount Log Winsorized'], bins=30, ax=ax[1], kde=True)
ax[1].set_title('Winsorized Total Claim Amount Log')

sns.histplot(df['Customer Lifetime Value Winsorized'], bins=30, ax=ax[2], kde=True)
ax[2].set_title('Winsorized Customer Lifetime Value')

plt.tight_layout()
plt.show()
```

## Enconding and Scaling Methodologies

1. **Ordinal Mapping** (for Coverage, Education, Marital Status, Employment Status):
   - These features likely have a natural order or ranking.
   - **Coverage**: Typically, coverage levels (like basic, extended, premium) have an inherent order from low to high.
   - **Education**: Educational levels usually follow a hierarchical order (e.g., High School < College < Masters).
   - **Marital Status**: If the model assumes a certain kind of order or impact based on marital status (e.g., single < married < divorced).
   - **Employment Status**: Employment types can be ordered based on stability or income potential (e.g., Unemployed < Employed < Retired).
   - By using ordinal encoding, we maintain the order which can be important for certain models to understand the intrinsic relationship between different categories.

2. **One-Hot Encoding** (for Vehicle Class, Renew Offer Type):
   - These categorical variables likely do not have any inherent order or ranking.
   - **Vehicle Class**: Different types of vehicles (e.g., sedan, SUV, sports car) don't have a natural order but are distinct categories.
   - **Renew Offer Type**: Different offer types are distinct and not ordinal.
   - One-hot encoding is used for nominal data where no ordinal relationship exists. It creates new binary columns, each representing a level in the original data.

3. **Robust Scaling** (for Number of Policies, Monthly Premium Auto, Total Claim Amount, Income, Policies Premium Interaction, Income Premium Ratio):
   - **Robust to Outliers**: Robust scaling is less sensitive to outliers in data. Given that your dataset likely contains financial metrics (like premiums and claims), which can often be skewed or have outliers, robust scaling is a good choice.
   - **Relative Scaling**: It scales the data according to the interquartile range, making it suitable for features with different ranges. This is useful for features like income and premiums which can vary significantly in scale.
   - **Policies Premium Interaction & Income Premium Ratio**: These seem to be derived features (interactions or ratios), and scaling them makes sense to keep their values within a comparable range with other features.

### Pipeline Preprocessing

```python
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from category_encoders import OrdinalEncoder as ce_OrdinalEncoder
from pycaret.internal.preprocess.transformers import TransformerWrapper

# Custom Transformer for Feature Creation
class FeatureCreation(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass 
    
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        # Feature 1: Interaction between Policies and Premiums
        X['Policies Premium Interaction'] = X['Number of Policies'] * X['Monthly Premium Auto']
        
        # Feature 2: Income to Premium Ratio
        X['Income Premium Ratio'] = X['Income'] / (X['Monthly Premium Auto'] + 0.01)

        # Feature 3: Demographics Score
        X['Demographics Score'] = (X['Marital Status'] + X['EmploymentStatus'] + X['Education']) / 3
        
        return X

# Mapping for OrdinalEncoder
ordinal_mapping = [
    {'col': 'Coverage', 'mapping': {'Basic': 1, 'Extended': 2, 'Premium': 3}},
    {'col': 'Education', 'mapping': {'High School or Below': 1, 'College': 2, 'Bachelor': 3, 'Master': 4, 'Doctor': 5}},
    {'col': 'Marital Status', 'mapping': {'Married': 2, 'Single': 1, 'Divorced': 0}},
    {'col': 'EmploymentStatus', 'mapping': {'Employed': 4, 'Unemployed': 0, 'Medical Leave': 2, 'Disabled': 1, 'Retired': 3}}
]

# Combine encoders in column transformer
encoder = ColumnTransformer([
    (
        'ordinal', ce_OrdinalEncoder(mapping=ordinal_mapping), 
        ['Coverage', 'Education', 'Marital Status', 'EmploymentStatus']
    ),
    (
        'onehot', OneHotEncoder(drop='first'), 
        ['Vehicle Class', 'Renew Offer Type']
    ),
], remainder='passthrough', verbose_feature_names_out=False)

# Combine scaler in column transformer
scaler = ColumnTransformer([
    (
        'scaler', RobustScaler(), 
        ['Number of Policies', 'Monthly Premium Auto', 'Total Claim Amount', 'Income', 'Policies Premium Interaction', 'Income Premium Ratio']
    ),
], remainder='passthrough', verbose_feature_names_out=False)

# Integrate all steps
pipeline = [
    ('encoder', TransformerWrapper(
        transformer=encoder)
    ),
    ('creator', TransformerWrapper(
        transformer=FeatureCreation())
    ),
    ('scaler', TransformerWrapper(
        transformer=scaler)
    ),
]
pipeline
```

## Pipeline Workflow

### Step 1: Custom Feature Creation
- The `FeatureCreation` class is defined as a custom transformer to create new features based on existing ones. 
- The `fit()` method does not alter anything in the data and returns the instance.
- The `transform()` method creates three new features:
    1. **Policies Premium Interaction**: Multiplies 'Number of Policies' with 'Monthly Premium Auto'.
    2. **Income Premium Ratio**: Calculates the ratio of 'Income' to ('Monthly Premium Auto' + 0.01) to prevent division by zero.
    3. **Demographics Score**: Averages the values of 'Marital Status', 'EmploymentStatus', and 'Education'.

### Step 2: Encoding Categorical Variables
- `ordinal_mapping` defines ordinal mappings for specific categorical columns.
- Two types of encoders are used within a `ColumnTransformer`:
    - **Ordinal Encoder**: Maps ordinal columns according to specified mappings.
    - **One-Hot Encoder**: Encodes nominal columns like 'Vehicle Class' and 'Renew Offer Type' into binary vectors.
- `remainder='passthrough'` preserves columns not subjected to encoding.
  
### Step 3: Scaling Numeric Variables
- `RobustScaler()` is used to scale numeric columns robustly against outliers within a `ColumnTransformer`.
- It only scales selected numeric columns while preserving the rest.

### Step 4: Integrating Steps into a Pipeline
- Three steps (`encoder`, `creator`, and `scaler`) are combined using `TransformerWrapper` instances.
- `TransformerWrapper` allows including non-sklearn transformers into the `sklearn` pipeline.

### Summary of Processing Steps:
1. **Encoding**: Transform categorical columns using `OrdinalEncoder` and `OneHotEncoder`.
2. **Feature Creation**: Generate new features using the `FeatureCreation` class.
3. **Scaling**: Scale selected numeric columns using `RobustScaler`.
4. **Pipeline Integration**: Combine all steps into a single pipeline for streamlined execution.
